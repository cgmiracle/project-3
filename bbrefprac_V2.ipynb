````python
margin = df['MARGIN']
assists = df['AST']
plt.scatter(margin, assists)
plt.xlabel('margin')
plt.ylabel('assists')
plt.show()
````

````python
import scipy.stats as st
cor = round(st.pearsonr(margin, assists)[0], 2)
print(f'Correleation between margin and assists is {cor}')
model = st.linregress(margin, assists)
y_values = margin * model[0] + model[1]
plt.scatter(margin, assists)
plt.plot(margin, y_values, color="red")
plt.xlabel('margin')
plt.ylabel('assists')
plt.show()
````
