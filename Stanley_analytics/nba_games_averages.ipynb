````python
home_away_df= pd.DataFrame(columns=away_df.columns)
#create empty dataframe
````

````python
for ind in playoffs_df.index:
    home_team_id = playoffs_df['HOME_TEAM_ID'][ind]
    away_team_id = playoffs_df['VISITOR_TEAM_ID'][ind]
    df2 = df_m.loc[df_m['HOME_TEAM_ID'] == home_team_id]
    df2 = df2.head(30)
    df2 = df2.reset_index(drop=True)
    average_home = df2.groupby(['HOME_TEAM_ID'])
    home_df = average_home.mean()
    home_df = home_df[['PTS_home', 'PTS_away', 'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home',
       'REB_home', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away',
       'AST_away', 'REB_away']]
    df3 = df_m.loc[df_m['VISITOR_TEAM_ID'] == away_team_id]
    df3 = df3.head(30)
    df3 = df3.reset_index(drop=True)
    average_away = df3.groupby(['VISITOR_TEAM_ID'])
    away_df = average_away.mean()
    away_df = away_df[['PTS_home', 'PTS_away', 'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home',
       'REB_home', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away',
       'AST_away', 'REB_away']]
    empty_ = []
    empty_.append(home_df.iloc[0,0])
    empty_.append(away_df.iloc[0,1])
    empty_.append(home_df.iloc[0,2])
    empty_.append(home_df.iloc[0,3])
    empty_.append(home_df.iloc[0,4])
    empty_.append(home_df.iloc[0,5])
    empty_.append(home_df.iloc[0,6])
    empty_.append(away_df.iloc[0,7])
    empty_.append(away_df.iloc[0,8])
    empty_.append(away_df.iloc[0,9])
    empty_.append(away_df.iloc[0,10])
    empty_.append(away_df.iloc[0,11])
    length_of_df = len(home_away_df)
    home_away_df.loc[length_of_df] = empty_
````

````python
home_away_df['HOME_TEAM_WINS'] = playoffs_df['HOME_TEAM_WINS']
````

````python
home_away_df
````

````python
home_away_df.to_csv('imported_predicted_df.csv', index=False)
````
